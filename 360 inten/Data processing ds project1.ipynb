```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
project = pd.read_csv(r"/content/Datasets.csv")
```

# Let's find outliers in Actual_Shipment_Time

```python
sns.boxplot(project.Actual_Shipment_Time)
```

# No outliers in Actual_Shipment_Time column
# Detection of outliers (find limits for Actual_Shipment_Time based on IQR)

```python
IQR = project['Actual_Shipment_Time'].quantile(0.75) - project['Actual_Shipment_Time'].quantile(0.25)
```

```python
lower_limit1 = project['Actual_Shipment_Time'].quantile(0.25) - (IQR * 1.5)
```

```python
upper_limit1 = project['Actual_Shipment_Time'].quantile(0.75) + (IQR * 1.5)
```

# Let's find outliers in Planned_Delivered_Time

```python
sns.boxplot(project.Planned_Delivery_Time)
```

# No outliers in Planned_Delivered_Time column

# Detection of outliers (find limits for Planned_Delivery_Time based on IQR)

```python
IQR = project['Planned_Delivery_Time'].quantile(0.75) - project['Planned_Delivery_Time'].quantile(0.25)
```

```python
lower_limit3 = project['Planned_Delivery_Time'].quantile(0.25) - (IQR * 1.5)
```

```python
upper_limit3 = project['Planned_Delivery_Time'].quantile(0.75) + (IQR * 1.5)
```

# Let's find outliers in Planned_Shipment_Time

```python
sns.boxplot(project.Planned_Shipment_Time)
```

# No outliers in Planned_Shipment_Time column

# Detection of outliers (find limits for Planned_Shipment_Time based on IQR)

```python
IQR = project['Planned_Shipment_Time'].quantile(0.75) - project['Planned_Shipment_Time'].quantile(0.25)
```

```python
lower_limit2 = project['Planned_Shipment_Time'].quantile(0.25) - (IQR * 1.5)
```

```python
upper_limit2 = project['Planned_Shipment_Time'].quantile(0.75) + (IQR * 1.5)
```

############### 1. Remove (let's trim the dataset) ################
# Trimming Technique
# Let's flag the outliers in the dataset

```python
outliers_project1 = np.where(project.Actual_Shipment_Time > upper_limit1, True, np.where(project.Actual_Shipment_Time < lower_limit1, True, False))
```

```python
# outliers data
project_out1 = project.loc[outliers_project1, ]
project_trimmed1 = project.loc[~(outliers_project1), ]
project.shape, project_trimmed1.shape
```

```python
# Let's explore outliers in the trimmed dataset
sns.boxplot(project_trimmed1.Actual_Shipment_Time)
```

```python
outliers_project2 = np.where(project.Planned_Shipment_Time > upper_limit1, True, np.where(project.Planned_Shipment_Time < lower_limit1, True, False))

# outliers data
project_out2 = project.loc[outliers_project2, ]
project_trimmed2 = project.loc[~(outliers_project2), ]
project.shape, project_trimmed2.shape
```

```python
# Let's explore outliers in the trimmed dataset
sns.boxplot(project_trimmed2.Actual_Shipment_Time)
```

```python
outliers_project3 = np.where(project.Planned_Delivery_Time > upper_limit1, True, np.where(project.Planned_Delivery_Time < lower_limit1, True, False))

# outliers data
project_out3 = project.loc[outliers_project3, ]
project_trimmed3 = project.loc[~(outliers_project3), ]
project.shape, project_trimmed3.shape
```

```python
# Let's explore outliers in the trimmed dataset
sns.boxplot(project_trimmed3.Actual_Shipment_Time)
```

############### 2. Replace ###############
# Replace the outliers by the maximum and minimum limit

```python
project['project_replaced1'] = pd.DataFrame(np.where(project['Actual_Shipment_Time'] > upper_limit1, upper_limit1, np.where(project['Actual_Shipment_Time'] < lower_limit1, lower_limit1, project['Actual_Shipment_Time'])))
sns.boxplot(project.project_replaced1)
```

```python
project['project_replaced2'] = pd.DataFrame(np.where(project['Planned_Shipment_Time'] > upper_limit2, upper_limit2, np.where(project['Planned_Shipment_Time'] < lower_limit2, lower_limit2, project['Planned_Shipment_Time'])))
sns.boxplot(project.project_replaced2)
```

```python
project['project_replaced3'] = pd.DataFrame(np.where(project['Planned_Delivery_Time'] > upper_limit1, upper_limit1, np.where(project['Planned_Delivery_Time'] < lower_limit1, lower_limit1, project['Planned_Delivery_Time'])))
sns.boxplot(project.project_replaced3)
```
